```python
class Status():
    def __init__(self, name, does_damage=None, prevents_damage=None,
                 is_preshop=None, is_postshop=None, is_ability=None, n_charges = 999):
        self.name = name
        self.does_damage = does_damage
        self.prevents_damage = prevents_damage
        self.is_preshop = is_preshop
        self.is_postshop = is_postshop
        self.is_ability = is_ability
        self.n_charges = n_charges
        
        if name == 'Garlic':
            self.prevents_damage = {'n': 2, 'full_shield': False}
            
        elif name == 'Lemon':
            self.prevents_damage = {'n': 4, 'full_shield': False}
            
        elif name == 'Melon':
            self.prevents_damage = {'n': 20, 'full_shield': True}
            self.n_charges = 1
            
        elif name == 'Coconut':
            self.prevents_damage = {'n': 999, 'full_shield': True}
            self.n_charges = 1
        
        elif name == 'Weak':
            self.prevents_damage = {'n': -3, 'full_shield': False}
        
        elif name == 'Meat Bone':
            self.does_damage = {'n': 4}
            self.n_charges = 1
        
        elif name == 'Steak':
            self.does_damage = {'n': 20}
            self.n_charges = 1
        
        elif name == 'Cheese':
            self.does_damage = {'mult': 2}
            self.n_charges = 1
        
        elif name == 'Fortune Cookie':
            self.does_damage = {'mult': 2, 'cond_func': lambda: np.random.choice([True, False])}
            self.n_charges = 999

        elif name == 'Honey':
            self.is_ability = {'trigger': 'Faint', 'bywhom': 'Self', 'trigger_cond': {},
                              'target': [Self], 'n_targets': -1, 'target_pos': (),
                              'effect': [Summon], 'effect_from': [NoTarget], 'effect_info': {'pet_func': HoneySummon},
                              'effect_repeat': 1}
            
        elif name == 'Mushroom':
            self.is_ability = {'trigger': 'Faint', 'bywhom': 'Self', 'trigger_cond': {},
                              'target': [Self], 'n_targets': -1, 'target_pos': (),
                              'effect': [Summon], 'effect_from': [NoTarget], 'effect_info': {'pet_func': MushroomSummon},
                              'effect_repeat': 1}
            
        elif name == 'Popcorns':
            self.is_ability = {'trigger': 'Faint', 'bywhom': 'Self', 'trigger_cond': {},
                              'target': [Self], 'n_targets': -1, 'target_pos': (),
                              'effect': [Summon], 'effect_from': [NoTarget], 'effect_info': {'pet_func': PopcornsSummon},
                              'effect_repeat': 1}
            
        elif name == 'Cucumber':
            self.is_ability = {'trigger': 'EndTurn', 'bywhom': 'Player', 'trigger_cond': {},
                               'target': [Self], 'n_targets': -1, 'target_pos': (),
                               'effect': [ModifyStats], 'effect_from': [NoTarget], 
                               'effect_info': {'add_hp': 1},'effect_repeat': 1}
            
        elif name == 'Croissant':
            self.is_ability = {'trigger': 'EndTurn', 'bywhom': 'Player', 'trigger_cond': {},
                               'target': [Self], 'n_targets': -1, 'target_pos': (),
                               'effect': [ModifyStats], 'effect_from': [NoTarget], 
                               'effect_info': {'add_atk': 1}, 'effect_repeat': 1}
            
        elif name == 'Carrot':
            self.is_ability = {'trigger': 'EndTurn', 'bywhom': 'Player', 'trigger_cond': {},
                               'target': [Self], 'n_targets': -1, 'target_pos': (),
                               'effect': [ModifyStats], 'effect_from': [NoTarget], 
                               'effect_info': {'add_atk': 1, 'add_hp': 1}, 'effect_repeat': 1}
            
        elif name == 'Grapes':
            self.is_ability = {'trigger': 'StartOfTurn', 'bywhom': 'Player', 'trigger_cond': {},
                               'target': [Self], 'n_targets': -1, 'target_pos': (),
                               'effect': [GainGold], 'effect_from': [NoTarget], 
                               'effect_info': {'gold': 1}, 'effect_repeat': 1}
        
    @property
    def n_charges(self):
        return self._n_charges
    @n_charges.setter
    def n_charges(self, new_val):
        self._n_charges = new_val
        if self._n_charges <= 0:
            self.name = 'Used ' + self.name
        
    def __repr__(self):
        return self.name
    
status_list = 'Melon, Coconut, Lemon, Garlic, Weak, Meat Bone, Steak, Cheese, Fortune Cookie, \
Peanut, Chili, Strawberry, Pineapple, Honey, Mushroom, Popcorns, NoStatus'.split(', ')
```
